```python
import pandas as pd
```

```python
waypoints = pd.read_csv('../data/osstate/waypoints_master.csv')
# migh require some manual edit: replace "ident,lat,lon" with "" and add a header "ident,lat,lon" again
waypoints.head()
```

```python
print(f'There are {waypoints.shape[0]} waypoints in the database')
```

```python
wp_lats = waypoints['lat'].values[:2000]
wp_lons = waypoints['lon'].values[:2000]
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Create a Cartopy PlateCarree projection
projection = ccrs.PlateCarree()

# Create a figure and axis with the PlateCarree projection
fig, ax = plt.subplots(subplot_kw={'projection': projection})

# Plot the points on the map
ax.plot(wp_lons, wp_lats, 'ro', transform=projection, markersize=1)

# Add map features
ax.coastlines()

# Set the extent of the map
ax.set_extent([-180, 180, -90, 90], crs=projection)

# Show the plot
plt.show()
```
